```python
import numpy as np
import pandas as pd
```

```python
df  = pd.read_csv("adult.csv")
df.head()
```

```python
m = df.shape[0]
print(f"No of training examples : {m}")
cols = df.shape[1]
print(f"No of input features are : {cols - 1}")
```

```python
df.info()
```

```python
missing_count = (df == "?").sum()
print(missing_count)
```

```python
categorical_cols = df.select_dtypes(include = ['object', 'category']).columns
print(f"No. of categorical column : {len(categorical_cols)} ")

for col in categorical_cols:
    print(f"{col} : {df[col].nunique()} unique categories")
```

```python
def replace_by_mode(column):
    mode_value = df.loc[df[column] != "?", column].mode()[0]  
    df.loc[df[column] == "?", column] = mode_value            
    return df[column]
```

```python
# replaceing with mode since no of unique categories are very less
df["workclass"] = replace_by_mode("workclass")
df["occupation"] = replace_by_mode("occupation")

# deleting native country with ? in entries
df["native.country"] = df["native.country"].replace("?", np.nan)
df = df.dropna(subset = ["native.country"])
```

```python
n = df.shape[0]

print(f"dropped training example : {m - n}")
```

```python
missing_count = (df == "?").sum()
print(missing_count)
```
