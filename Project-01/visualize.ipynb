```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_adult.csv")
```

```python
df.head(2)
```

```python
def stat(column):
    series = df[column]   # work on a single column
    return f"""
        Feature: {column}
        Mean: {series.mean()}
        Median: {series.median()}
        Mode: {series.mode().iloc[0]}
        Std Dev: {series.std()}
        Variance: {series.var()}
        """
```

```python
numerical_cols =df.select_dtypes(include = ["int64", "float64"])

for col in numerical_cols:
    print(col)
```

```python
print(stat("age"))
print(stat("fnlwgt"))
print(stat("education.num"))
```

```python
print(stat("capital.gain"))
print(stat("capital.loss"))
print(stat("hours.per.week"))
```

```python
# AGE:
# hist for age : shows age distribution of people in dataset
df["age"].hist(bins = 20)
plt.title("Age Distribution")
plt.show()
```

```python
# OCCUPATION
# this bar shows the distribution of occupation, among people in the given data]
df["occupation"].value_counts().plot(kind = "bar")
plt.title("occupation")
plt.show()
```

```python
# MARITIAL STATUS
# this bar visualize the marital.status  status of people
df["marital.status"].value_counts().plot(kind = "bar")
plt.title("marital.status")
plt.show()
```

```python
# EDUATION
df["education.num"].hist(bins = 10)
plt.xlabel("Years of eduation")
plt.ylabel("no of ppl")
plt.title("Eduation")
plt.show()
```

```python
# income
df["income"].value_counts().plot(kind = "bar")
plt.title("Income")
plt.show()
```

```python
sns.boxplot(x="age", y="income", data=df)
plt.title("income Distribution by Age Category")
plt.show()
```

```python
# capital gain vs income
sns.boxplot(x = "capital.gain" , y = "income", data = df)
plt.title("capital gain vs income")
plt.show()
```

```python
# occupation vs income.
crosstab = pd.crosstab(df["occupation"], df["income"])

# Plot stacked bar
crosstab.plot(kind="bar", stacked=True, figsize=(12,6))
plt.title("Occupation vs Income")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.legend(title="Income")
plt.xticks(rotation=45, ha="right")
plt.show()
```
